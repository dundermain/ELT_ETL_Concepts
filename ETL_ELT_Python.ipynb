1. Extract:
* Data Source: Identifies the source systems where the data resides (databases, files, APIs, etc.).
* Connection: Establishes a connection to the source system to access the data.
* Extraction: Retrieves the relevant data from the source system based on defined criteria.


2. Transform:
* Data Cleaning: Removes inconsistencies, errors, and duplicates from the extracted data.
* Data Validation: Ensures data accuracy and completeness against predefined rules.
* Data Standardization: Converts data into a consistent format (e.g., data types, units, encoding).
* Data Enrichment: Adds context or value to the data by combining it with other data sources.
* Data Aggregation: Combines data from multiple sources or summarizes data for analysis.


3. Load:

* Target System: Defines the destination system where the transformed data will be stored (data warehouse, data lake, database).
* Data Loading: Transfers the transformed data to the target system.
* Error Handling: Manages any errors or exceptions that occur during the loading process.


Additional Components:

* Data Profiling: Analyzes the structure, content, and quality of the source data to inform the transformation process.
* Metadata Management: Stores information about the data, its sources, transformations, and usage.
* Scheduling: Automates the execution of the ETL pipeline at specific intervals.
* Monitoring and Logging: Tracks the pipeline's performance, identifies issues, and records activities.



```python
#sample ETL pipeline

def extract(file_name):
  return pd.read_csv(file_name)

def transform(data_frame):
  return data_frame.loc[:, ["industry_name", "number_of_firms"]]

def load(data_frame, file_name):
  data_frame.to_csv(file_name)

extracted_data = extract(file_name="raw_industry_data.csv")
transformed_data = transform(data_frame=extracted_data)

# Pass the transformed_data DataFrame to the load() function
load(data_frame=transformed_data, file_name="number_of_firms.csv")
```
